# üß± Data model

```python
from dataclasses import dataclass, field
from typing import List, Optional, Dict

@dataclass
class Intention:
    finalidade: str  # 'analise','criacao','sintese','engenharia_reversa','interpretacao'
    rigidez: float   # 0.0 - 1.0
    dimensao: List[str]  # ['CoT','RoT',...]
    sd_alvo: float
    ambiguity_max: float

@dataclass
class Structure:
    tipo: str  # 'Tensorial','Fluxo','Curvatura',...
    descricao: Optional[str] = None

@dataclass
class Mechanisms:
    use_CoT: bool = False
    use_ToT: bool = False
    use_GoT: bool = False
    use_RoT: bool = False
    use_ER_OE: bool = False
    tensionadores: List[str] = field(default_factory=list)
    meta_audit: bool = True

@dataclass
class Persona:
    role_name: Optional[str] = None
    anchors: List[str] = field(default_factory=list)
    polarities: Dict[str,float] = field(default_factory=dict)
    horizon_consistency: int = 3

@dataclass
class ConstructorConfig:
    intention: Intention
    structure: Structure
    mechanisms: Mechanisms
    persona: Optional[Persona] = None
    output_format: str = "markdown"  # or 'json','md+code'
```

# C√©lula 2 ‚Äî Imports & helpers

# C√©lula 3 ‚Äî Defini√ß√µes (dataclasses) ‚Äî colar o bloco acima

```python
# ============================================================
# üß¨ C√âLULA 3 ‚Äî DEFINI√á√ïES NUCLEARES (MODELOS DE DADOS)
# ============================================================
# Esta c√©lula cont√©m APENAS:
# - dataclasses fundamentais
# - modelos cognitivos
# - estrutura interna do constructor
#
# N√ÉO busca arquivos.
# N√ÉO depende de templates externos.
# Tudo √© autocontido e robusto para qualquer LLM.
# ============================================================

from dataclasses import dataclass, field
from typing import List, Optional, Dict

# -----------------------------
# INTEN√á√ÉO
# -----------------------------
@dataclass
class Intention:
    finalidade: str            # 'analise','criacao','sintese','engenharia_reversa','interpretacao'
    rigidez: float             # 0.0 a 1.0
    dimensao: List[str]        # ['CoT','ToT','GoT','RoT']
    sd_alvo: float             # alvo de densidade sem√¢ntica
    ambiguity_max: float       # permiss√£o de ambiguidade residual

# -----------------------------
# ESTRUTURA COGNITIVA
# -----------------------------
@dataclass
class Structure:
    tipo: str                  # 'Tensorial','Fluxo','Curvatura','Estratifica√ß√£o','Superposi√ß√£o','Ruptura','Din√¢mica','Geod√©sica'
    descricao: Optional[str] = None


# -----------------------------
# MECANISMOS OPERACIONAIS
# -----------------------------
@dataclass
class Mechanisms:
    use_CoT: bool = False
    use_ToT: bool = False
    use_GoT: bool = False
    use_RoT: bool = False
    use_ER_OE: bool = False
    tensionadores: List[str] = field(default_factory=list)
    meta_audit: bool = True   # auditoria cognitiva no final


# -----------------------------
# PERSONA (ATRATOR SEM√ÇNTICO)
# -----------------------------
@dataclass
class Persona:
    role_name: Optional[str] = None
    anchors: List[str] = field(default_factory=list)
    polarities: Dict[str, float] = field(default_factory=dict)
    horizon_consistency: int = 3   # quantos turns mant√©m coer√™ncia


# -----------------------------
# CONFIGURA√á√ÉO GERAL DO CONSTRUCTOR
# -----------------------------
@dataclass
class ConstructorConfig:
    intention: Intention
    structure: Structure
    mechanisms: Mechanisms
    persona: Optional[Persona] = None
    output_format: str = "markdown"   # 'markdown', 'json', 'md+code'
```

# C√©lula 4 ‚Äî Fun√ß√µes de constru√ß√£o (core)

```python
def build_system_instruction(config: ConstructorConfig) -> str:
    # cria cabe√ßalho system prompt
    i = config.intention
    s = config.structure
    mech = config.mechanisms
    persona = config.persona

    lines = []
    lines.append(f"SYSTEM: Voc√™ √© um agente especializado em '{i.finalidade}'.")
    lines.append(f"Objetivo: produzir sa√≠da com densidade sem√¢ntica alvo SD ‚â• {i.sd_alvo:.2f} e ambiguidade ‚â§ {i.ambiguity_max:.2f}.")
    lines.append(f"Grau de rigidez operacional: {i.rigidez:.2f} (0=explorar,1=strito).")
    lines.append(f"Estrutura Cognitiva: {s.tipo} ‚Äî {s.descricao or ''}")
    lines.append("Mecanismos ativados:")
    mech_lines = []
    for flag, name in [ (mech.use_CoT, "CoT"), (mech.use_ToT, "ToT"),
                        (mech.use_GoT, "GoT"), (mech.use_RoT, "RoT"),
                        (mech.use_ER_OE, "ER-OE")]:
        if flag:
            mech_lines.append(f"- {name}")
    if mech.tensionadores:
        mech_lines.append(f"- Tensionadores: {', '.join(mech.tensionadores)}")
    lines.extend(mech_lines)
    if persona:
        lines.append(f"Persona (atrator sem√¢ntico): {persona.role_name}. √Çncoras: {persona.anchors[:5]}")
        lines.append(f"Horizonte de consist√™ncia: {persona.horizon_consistency} turns")
    lines.append("Contrato de sa√≠da: Responda estritamente no formato especificado em 'Output Format'.")
    return "\n".join(lines)

def build_process_block(config: ConstructorConfig, task_description: str) -> str:
    # passo-a-passo de execu√ß√£o
    mech = config.mechanisms
    blocks = []
    blocks.append("PROCESSO (siga estritamente):")
    blocks.append("1) Step-Back: Resuma em 1 frase o princ√≠pio abstrato subjacente √† tarefa.")
    if mech.use_ToT:
        blocks.append("2) ToT: Gere 3 paths (pensamentos) distintos; para cada path, gere 2 passos subsequentes; avalie viabilidade.")
    if mech.use_GoT:
        blocks.append("3) GoT: Construa grafo de depend√™ncias entre conceitos chave.")
    if mech.use_CoT:
        blocks.append("4) CoT: Para o path selecionado, gere racioc√≠nio passo-a-passo (explique por que cada passo).")
    if mech.use_RoT:
        blocks.append("5) RoT: Execute ciclo de reflex√£o ‚Äî avalie coer√™ncia, identifique contrad. se houver, corrija e reexecute.")
    if mech.use_ER_OE:
        blocks.append("6) ER-OE: Desmonte hip√≥teses, extraia m√©tricas e proponha experimentos reproduz√≠veis.")
    if mech.meta_audit:
        blocks.append("7) Auditoria final: Verifique SD, Ambiguidade, e insira um relat√≥rio curto de conformidade (3 linhas).")
    blocks.append(f"Task: {task_description}")
    return "\n".join(blocks)

def build_output_spec(config: ConstructorConfig) -> str:
    fmt = config.output_format
    return f"OUTPUT FORMAT: {fmt}\nSe {fmt} == markdown: Incluir YAML header com {json.dumps(asdict(config), indent=2)}\nIncluir se√ß√£o: Resumo Executivo (1-3 frases), Componentes T√©cnicos, Passos Reprodut√≠veis, Refer√™ncias."
```

# C√©lula 5 ‚Äî Instanciar config (exemplo)

```python
# exemplo de configura√ß√£o
i = Intention(finalidade='engenharia_reversa', rigidez=0.8,
              dimensao=['CoT','RoT'], sd_alvo=0.85, ambiguity_max=0.18)
s = Structure(tipo='Curvatura', descricao='Foco em proje√ß√µes geod√©sicas no espa√ßo sem√¢ntico.')
m = Mechanisms(use_CoT=True, use_ToT=True, use_GoT=False, use_RoT=True, use_ER_OE=True,
               tensionadores=['oposto_complementar','padr√£o_probabilidade'], meta_audit=True)
p = Persona(role_name='Reverse-Scientist', anchors=['seguran√ßa','reprodutibilidade','matriz'], polarities={'serio':0.9}, horizon_consistency=5)

config = ConstructorConfig(intention=i, structure=s, mechanisms=m, persona=p, output_format='markdown')
```

# C√©lula 6 ‚Äî Gerar prompt mestre

```python
system_instr = build_system_instruction(config)
process_block = build_process_block(config, "Extra√ß√£o cient√≠fica: 'Causal Tracing em Transformers'")
output_spec = build_output_spec(config)

prompt_mestre = f"""{system_instr}

CONTEXTO: Voc√™ √© ativado para uma investiga√ß√£o cient√≠fica com requisitos de rigor e reproducibilidade.

{process_block}

{output_spec}

INSTRU√á√ïES ADICIONAIS:
- Use linguagem t√©cnica objetiva, sem floreios.
- Sempre inclua evid√™ncia (ex.: refer√™ncia arXiv/DOI/GitHub) quando fizer claims.
- Se n√£o souber algo, declare limitacao explicitamente.
- Ao final, execute RoT: 'Revis√£o cr√≠tica em 3 frases' e um 'Checklist SD/Ambiguidade'.

RESPONDA AGORA:
"""
print(prompt_mestre[:1200])
```

# üìå C√©lula X ‚Äî Helpers de Valida√ß√£o (Quality Gates autom√°ticos)

```python
# C√©lula: Quality gates / valida√ß√µes r√°pidas
from typing import Any

def validate_config(cfg: ConstructorConfig) -> List[str]:
    errs = []
    # dimens√£o cont√©m pelo menos um m√©todo
    if not cfg.intention.dimensao or not any(d in ['CoT','ToT','GoT','RoT'] for d in cfg.intention.dimensao):
        errs.append("Intention.dimensao deve conter pelo menos 'CoT'|'ToT'|'GoT'|'RoT'.")
    # sd_alvo para tarefas cient√≠ficas
    if cfg.intention.sd_alvo < 0.0 or cfg.intention.sd_alvo > 1.0:
        errs.append("sd_alvo deve estar em [0.0, 1.0].")
    # sugerir m√≠nimo para tasks cient√≠ficas
    if cfg.intention.finalidade in ('engenharia_reversa','analise','sintese') and cfg.intention.sd_alvo < 0.7:
        errs.append("sd_alvo < 0.7 ‚Äî para tarefas cient√≠ficas recomendo sd_alvo >= 0.7.")
    # ambiguity
    if cfg.intention.ambiguity_max < 0.0 or cfg.intention.ambiguity_max > 1.0:
        errs.append("ambiguity_max deve estar em [0.0, 1.0].")
    if cfg.intention.ambiguity_max > 0.25:
        errs.append("ambiguity_max alto (>0.25). Para rigor cient√≠fico recomendo <= 0.18.")
    # output format contract
    if cfg.output_format not in ('markdown','json','md+code'):
        errs.append("output_format deve ser 'markdown', 'json' ou 'md+code'.")
    # persona consistency
    if cfg.persona:
        if cfg.persona.horizon_consistency < 1:
            errs.append("persona.horizon_consistency deve ser >= 1.")
    return errs

def run_quality_checks_and_report(cfg: ConstructorConfig) -> None:
    errs = validate_config(cfg)
    print("=== Quality Gates Report ===")
    if not errs:
        print("OK ‚Äî configura√ß√£o atende √†s checks b√°sicas.")
    else:
        print("ERROS / AVISOS:")
        for e in errs:
            print(f" - {e}")
    # Quick sanity stats
    print("\nResumo r√°pido de par√¢metros:")
    print(f" Finalidade: {cfg.intention.finalidade}")
    print(f" Dimens√µes: {cfg.intention.dimensao}")
    print(f" SD alvo: {cfg.intention.sd_alvo:.2f}")
    print(f" Ambiguidade m√°xima: {cfg.intention.ambiguity_max:.2f}")
    print(f" Estrutura: {cfg.structure.tipo}")
    print(f" Mecanismos ativados: " +
          ", ".join([n for f,n in [(cfg.mechanisms.use_CoT,'CoT'),
                                   (cfg.mechanisms.use_ToT,'ToT'),
                                   (cfg.mechanisms.use_GoT,'GoT'),
                                   (cfg.mechanisms.use_RoT,'RoT'),
                                   (cfg.mechanisms.use_ER_OE,'ER-OE')] if f]))
    print(f" Tensionadores: {cfg.mechanisms.tensionadores}")
    print("=============================\n")
```

# üìå C√©lula Z ‚Äî Fun√ß√µes utilit√°rias: salvar prompt, exportar YAML header, imprimir resum√£o

```python
# C√©lula: salvar prompt, exportar metadata em YAML e print resumido
import yaml
from datetime import datetime

def save_prompt_to_file(prompt_text: str, cfg: ConstructorConfig, filename: str):
    # escreve prompt e header YAML num √∫nico arquivo .md
    md_header = {
        'config': asdict(cfg),
        'generated_at': datetime.utcnow().isoformat() + 'Z'
    }
    content = f"---\n{yaml.safe_dump(md_header, sort_keys=False)}---\n\n"
    content += prompt_text
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(content)
    print(f"Prompt salvo em: {filename}")

def summarize_and_print(prompt_text: str, max_chars:int=1200):
    print("=== Prompt (preview) ===\n")
    print(prompt_text[:max_chars])
    if len(prompt_text) > max_chars:
        print("\n... (preview truncado) ...")
    print("\n========================\n")
```

# üìå C√©lula W ‚Äî Test harness: executar checks e salvar prompt mestre
